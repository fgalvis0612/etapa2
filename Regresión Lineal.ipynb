```python
#Instalar las librerias
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
import os
```

```python
#Cargar los datos
data = pd.read_csv("C:/Users/Felipe Galvis/Documents/Python/data.csv")
```

```python
#Mostrar los datos
data.head(5)
```

```python
#Grafica de dispersion
data.plot.scatter(x="metro",y="precio")
plt.show()
```

```python
#Regresion Lineal
regresion = linear_model.LinearRegression()

#Agregar los datos en un array
metros = data["metro"].values.reshape((-1,1))

#modelo
modelo = regresion.fit(metros, data["precio"])

print("Interaccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
#Mostrar segun los siguientes valores
entrada = [[5],[10],[15],[25]]
prediccion = modelo.predict(entrada)
print(prediccion)
```

```python
#Linea de regresion
data.plot.scatter(x="metro", y="precio", label='Datos originales')
plt.scatter(entrada, prediccion, color='red')
plt.plot(entrada,prediccion, color='black', label='Linea de regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
